You are given an integer, $N$. Your task is to print an alphabet rangoli of size $N$. (Rangoli is a form of Indian folk art based on creation of patterns.)

Different sizes of alphabet rangoli are shown below:

<code>#size 3
----c----
--c-b-c--
c-b-a-b-c
--c-b-c--
----c----</code>

<code>#size 5
--------e--------

e-d-c-b-a-b-c-d-e
--e-d-c-b-c-d-e--
----e-d-c-d-e----
------e-d-e------
--------e--------</code>

<code>#size 10
------------------j------------------
----------------j-i-j----------------
--------------j-i-h-i-j--------------
------------j-i-h-g-h-i-j------------
----------j-i-h-g-f-g-h-i-j----------
--------j-i-h-g-f-e-f-g-h-i-j--------

----j-i-h-g-f-e-d-c-d-e-f-g-h-i-j----
--j-i-h-g-f-e-d-c-b-c-d-e-f-g-h-i-j--
j-i-h-g-f-e-d-c-b-a-b-c-d-e-f-g-h-i-j

--------------j-i-h-i-j--------------
----------------j-i-j----------------
------------------j------------------</code>

The center of the rangoli has the first alphabet letter a, and the boundary has the $N^{th}$ alphabet letter (in alphabetical order).

<b>Function Description</b>

Complete the rangoli function in the editor below.

rangoli has the following parameters:

- int size: the size of the rangoli

<b>Returns</b>

- string: a single string made up of each of the lines of the rangoli separated by a newline character (\n)

<b>Input Format</b>

Only one line of input containing $size$, the size of the rangoli.

<b>Constraints</b>

$0 \leq size \leq 27$

<b>Sample Input</b>

<code>     5</code>

<b>Sample Output</b>

<code>
--------e--------
------e-d-e------
----e-d-c-d-e----
--e-d-c-b-c-d-e--
e-d-c-b-a-b-c-d-e
--e-d-c-b-c-d-e--
----e-d-c-d-e----
------e-d-e------
--------e--------</code>

---

## Solution.

Create a list with each item from the bottom of the alphabet rangoli. Then join it with the same list in reverse order to obtain the figure.

```python
def print_rangoli(size):
    # your code goes here
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    mylist = []
    
    # "mylist" is a list that contains lower part of rangoli alphabet
    for i in range(size):
        # Select the letter of the alphabet and use hyphen to join it
        s = "-".join(alphabet[i:size])
        
        # "s[::-1]" String in reverse order.
        mylist.append((s[::-1] + s[1:]).center(4*size-3, "-"))

    print('\n'.join(mylist[:0:-1] + mylist))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)    
```
