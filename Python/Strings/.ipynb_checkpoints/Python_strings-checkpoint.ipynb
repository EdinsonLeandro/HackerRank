### String Formatting

```python
def print_formatted(number):
    
    size = len(bin(number)[2:])
    
    for item in range(1,number + 1):
        print('{}'.format(int(item)).rjust(size),
              '{}'.format(oct(item)[2:]).rjust(size),
              '{}'.format(hex(item)[2:].upper()).rjust(size),
              '{}'.format(bin(item)[2:]).rjust(size))

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

### Alphabet Rangoli

```python
def print_rangoli(size):
    # your code goes here
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    L = []
    
    for i in range(size):
        s = "-".join(alphabet[i:size])
        L.append((s[::-1]+s[1:]).center(4*size-3, "-"))
        
    print('\n'.join(L[:0:-1]+L))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

```python
# initializing string  
def count_subs_in_string(string,substring):
    count = 0
    number = len(substring)

    for item in range(len(string) - (number-1)): 

        if substring in string[0 + item : number + item]: 
            count = count + 1
    return count

mylist = []
mydict = {'Stuart':0 ,'Kevin':0}

if word[0] in consonants:
    mydict['Stuart'] = 1
else:
    mydict['Kevin'] = 1


for element in range(len(word)-1):
    for item in range(len(word)-element):
        subword = word[0 + item : element + 1 + item]

        if subword not in mylist:
            mylist.append(subword)
            
            if subword[0] in consonants:
                mydict['Stuart'] += count_subs_in_string(word,subword)
            else:
                mydict['Kevin'] += count_subs_in_string(word,subword)

if mydict['Stuart'] > mydict['Kevin']:
    print('Stuart', mydict['Stuart'])

elif mydict['Stuart'] < mydict['Kevin']:
    print('Kevin', mydict['Kevin'])
    
```

### Merge the Tools!

```python
def merge_the_tools(string, k):
    n = len(string)
    factor = int(n/k) 

    for item in range(k):
        unique_str = ''
        sub_string = string[0 + factor*item : factor + factor*item]

        for item in sub_string: 
            if item not in unique_str: 
                unique_str += item

        print(unique_str)

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```
