Kevin and Stuart want to play the **'The Minion Game'**.

**Game Rules**

Both players are given the same string, $S$.

Both players have to make substrings using the letters of the string $S$.

Stuart has to make words starting with consonants.

Kevin has to make words starting with vowels.

The game ends when both players have made all possible substrings.

**Scoring**

A player gets +1 point for each occurrence of the substring in the string $S$.

**For Example:**

String $S$ = BANANA

Kevin's vowel beginning word = ANA

Here, ANA occurs twice in BANANA. Hence, Kevin will get 2 Points.

For better understanding, see the image below:



Your task is to determine the winner of the game and their score.

<b>Function Description</b>

Complete the minion_game in the editor below.

minion_game has the following parameters:

- string string: the string to analyze

<b>Prints</b>

- string: the winner's name and score, separated by a space on one line, or <code>Draw</code> if there is no winner.

<b>Input Format</b>

A single line of input containing the string $S$.

**Note:** The string $S$ will contain only uppercase letters: $[A - Z]$.

<b>Constraints</b>

$0 < len(S) \leq 10^6$

<b>Sample Input</b>

<code>     BANANA</code>

<b>Sample Output</b>

<code>     Stuart 12</code>

**Note:**

Vowels are only defined as $AEIOU$. In this problem, $Y$ is not considered a vowel.

---

## Solution.

Let's use <code>BANANA</code> example. Let's also see the position of each letter in the word <code>BANANA</code> inside the square brackets: <code>B[0] A[1] N[2] A[3] N[4] A[5]</code>.

It is important to see that all possible substrings starting with each letter are the number of letters in the word ("BANANA" has 6) minus the position of the letter within the string. For example:

- Starting with letter <code>B</code>, it means $6 - 0 = 6$.

- Starting with letter <code>A</code>, it means $6 - 1 = 5$.

And so on.

```python
def minion_game(word):
    vowels = 'AEIOU'
    
    # Initialize Kevin and Stuart points
    kevin_point = 0
    stuart_point = 0
    
    # Take each letter of the input word. If the letter is a vowel, calculate all possible substrings.
    # It the letter is vowel, sum to Kevin punctuation. Otherwise, sum to Stuart.
    for i in range(len(word)):
        all_substrings = len(word)-i
        if word[i] in vowels:
            kevin_point += all_substrings    
        else:
            stuart_point += all_substrings

    if kevin_point > stuart_point:
        print("Kevin", kevin_point)
        
    elif kevin_point < stuart_point:
        print("Stuart", stuart_point)
        
    else:
        print("Draw")

if __name__ == '__main__':
    s = input()
    minion_game(s)        
```
