### sWAP cASE

```python
def swap_case(s):
    result = ''
    for item in s:
        if item.isupper():
            result += item.lower()
        else:
            result += item.upper()

    return result

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
def count_substring(string, sub_string):
    mylist = []
    for item in range(len(string)):
        mylist.append(string[item:].find(sub_string))
        
    return mylist.count(0)

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

### String Formatting

```python
def print_formatted(number):
    
    size = len(bin(number)[2:])
    
    for item in range(1,number + 1):
        print('{}'.format(int(item)).rjust(size),
              '{}'.format(oct(item)[2:]).rjust(size),
              '{}'.format(hex(item)[2:].upper()).rjust(size),
              '{}'.format(bin(item)[2:]).rjust(size))

if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

### Alphabet Rangoli

```python
def print_rangoli(size):
    # your code goes here
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    L = []
    
    for i in range(size):
        s = "-".join(alphabet[i:size])
        L.append((s[::-1]+s[1:]).center(4*size-3, "-"))
        
    print('\n'.join(L[:0:-1]+L))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```
