### Nested Lists

```python
#names = ['Harry', 'Berry', 'Tina', 'Akriti', 'Harsh']
#scores = [37.21, 37.21, 37.2, 41, 39]
#names = ['Harsh', 'Beria', 'Varun', 'Kakunami', 'Vikas']
#scores = [20, 21, 19, 19, 19]
names = ['Harsh', 'Beria', 'Varun', 'Kakunami', 'Vikas']
scores = [20, 21, 19, 19, 21]
students = []

for item in range(len(names)):
    name = names[item]
    score = scores[item]
    students.append([name,score])

# take second element for sort
def second_element(elem):
    return elem[1]

def first_element(elem):
    return elem[0]

# sort list with key
students_sort = sorted(students,key=second_element)

#Determine the position of second lowest
for item in range(len(students_sort)-1):
    if students_sort[item][1] != students_sort[item+1][1]:
        result = item+1
        break   

preeliminary = [students_sort[result]]

mylist = [preeliminary.append(students_sort[i+1]) for i in range(result,len(students_sort)-1) if (students_sort[i][1] == students_sort[i+1][1] and students_sort[i]==preeliminary[0])]

result_sorted = sorted(preeliminary,key=first_element)

for item in result_sorted:
    print(item[0])
```

###  itertools.permutations()

```python
from itertools import permutations

print(list(permutations(['1','2','3'],2)))
```

```python
if __name__ == '__main__':
    string, k = input().split()
    k = int(k)
    string = sorted(string)
    
    [print(''.join(item)) for item in list(permutations(string,k))]
```

### itertools.combinations()

```python
from itertools import combinations

if __name__ == '__main__':
    string, k = input().split()
    k = int(k)
    string = sorted(string)
    
    for element in range(1, k+1):
        [print(''.join(item)) for item in list(combinations(string,element))]
```
