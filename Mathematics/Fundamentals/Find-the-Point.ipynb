## Problem.

Consider two points, $p = (p_x, p_y)$ and $q = (q_x, q_y)$. We consider the inversion or <b>point reflection</b>, $r = (r_x, r_y)$, of point $p$ across point $q$ to be a $180$ rotation of point $p$ around $q$.

Given $n$ sets of points $p$ and $q$, find $r$ for each pair of points and print two space-separated integers denoting the respective values of $r_x$ and $r_y$ on a new line.

<b>Function Description</b>

Complete the findPoint function in the editor below.

findPoint has the following parameters:

- int px, py, qx, qy: x and y coordinates for points $p$ and $q$.

<b>Returns</b>

- int[2]: x and y coordinates of the reflected point $r$.

<b>Input Format</b>

The first line contains an integer, $n$, denoting the number of sets of points.

Each of the subsequent lines contains four space-separated integers that describe the respective values of $p_x$, $p_y$, $q_x$, and $q_y$ defining points $p = (p_x, p_y)$ and $q = (q_x, q_y)$.

<b>Constraints</b>
- $1 \leq n \leq 15$
- $-100 \leq p_x, p_y, q_x, q_y \leq 100$


<b>Sample Input</b>

<code>2<br>0 0 1 1<br>1 1 2 2</code>


<b>Sample Output</b>

<code>2 2<br>3 3</code>

## Solution.

There is a simple way to compute the relationship between the coordinates of point reflection $r = (r_x, r_y)$ with the pair or points $p$ and $q$.

In the graph above, you can see that both $x$ and $y$ coordinates of point $r$ is twice the magnitude of the corresponding axis of point $q$ minus the value of point $p$.

```python
def findPoint(px, py, qx, qy):
    # print(rx, ry)
    print(qx + qx - px , qy + qy - py)

if __name__ == '__main__':

    n = int(input())

    for n_itr in range(n):
        pxPyQxQy = input().split()

        px = int(pxPyQxQy[0])

        py = int(pxPyQxQy[1])

        qx = int(pxPyQxQy[2])

        qy = int(pxPyQxQy[3])

        result = findPoint(px, py, qx, qy)
```
