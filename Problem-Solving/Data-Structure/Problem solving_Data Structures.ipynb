```python
# Complete the hourglassSum function below.
# Complete the hourglassSum function below.
def hourglassSum(arr):
    n_rows = len(arr) - 2
    n_cols = len(arr[0]) - 2
    result = []

    for item in range(n_rows):
        for el in range(n_cols):
            prev = sum(arr[item][0 + el : 3 + el]) + arr[item + 1][el + 1] + sum(arr[item + 2][0 + el : 3 + el])
            result.append(prev)
    
    return max(result)

if __name__ == '__main__':

    arr = []
    for _ in range(6):
        arr.append(list(map(int, input().rstrip().split())))

    result = hourglassSum(arr)
    print(result)
```

### Dynamic Array

```python
def dynamicArray(n, queries):
    lastAnswer = 0
    seqList = [[] for _ in range(n)]

    for item in range(q):
        x = queries[item][1]
        y = queries[item][2]
        seq = ((x ^ lastAnswer) % n)
        
        if queries[item][0] == 1:
            seqList[seq].append(y)
        else:
            mod_size = y % len(seqList[seq])
            lastAnswer = seqList[seq][mod_size]
            print(lastAnswer)

if __name__ == '__main__':

    first_multiple_input = input().rstrip().split()

    n = int(first_multiple_input[0])

    q = int(first_multiple_input[1])

    queries = []

    for _ in range(q):
        queries.append(list(map(int, input().rstrip().split())))

    result = dynamicArray(n, queries)
```

### Left Rotation

```python
def left_rotation(a,d):
    # a: List with elements to be rotated
    # d: number of  rotations
    
    for item in range(d):
        number_rotated = a.pop(0)
        a.append(number_rotated)
    
    return a
```

```python
if __name__ == '__main__':
    nd = input().split()

    n = int(nd[0])

    d = int(nd[1])

    a = list(map(int, input().rstrip().split()))

    mylist = left_rotation(a,d)

    print(*mylist, sep=' ')
```

### Array Manipulation

```python
if __name__ == '__main__':
    nm = input().split()

    n = int(nm[0])

    m = int(nm[1])

    queries = []

    for _ in range(m):
        queries.append(list(map(int, input().rstrip().split())))
```

```python
queries
```

```python
def arrayManipulation(n, queries):
    result = [0]*n
    
    # Number of operations:
    m = len(queries)
    
    for i in range(m):
        # Read left index, right index and summand (a, b, k)
        a, b, k = queries[i][0], queries[i][1], queries[i][2]

        result[a-1 : b] = map(lambda x: x + k, result[a-1 : b])

    print(max(result))
    print(result)
```

```python
arrayManipulation(5, queries)
```
