Maria plays college basketball and wants to go pro. Each season she maintains a record of her play. She tabulates the number of times she breaks her season record for most points and least points in a game. Points scored in the first game establish her record for the season, and she begins counting from there.

**Example**

$scores = [12, 24, 10, 24]$

Scores are in the same order as the games played. She tabulates her results as follows:

<code>                                     Count
    Game  Score  Minimum  Maximum   Min Max
     0      12     12       12       0   0
     1      24     12       24       0   1
     2      10     10       24       1   1
     3      24     10       24       1   1</code>
     
Given the scores for a season, determine the number of times Maria breaks her records for most and least points scored during the season.

<b>Function Description</b>

Complete the breakingRecords function in the editor below.

breakingRecords has the following parameter(s):

- int scores[n]: points scored per game.

<b>Returns</b>

- int[2]: An array with the numbers of times she broke her records. Index $0$ is for breaking most points records, and index $1$ is for breaking least points records.

<b>Input Format</b>

- The first line contains an integer $n$, the number of games.
- The second line contains $n$ space-separated integers describing the respective values of $scores_0, scores_1, ..., scores_{n-1}$.

<b>Constraints</b>

- $1 \leq n \leq 1000$
- $0 \leq scores[i] \leq 10^8$

<b>Sample Input 0</b>

<code>     9
     10 5 20 20 4 5 2 25 1</code>

<b>Sample Output</b>

<code>     2 4</code>

<b>Explanation 0</b>

The diagram below depicts the number of times Maria broke her best and worst records throughout the season:



She broke her best record twice (after games $2$ and $7$) and her worst record four times (after games $1$, $4$, $6$, and $8$), so we print <code>2 4 as</code> our answer. Note that she did not break her record for best score during game $3$, as her score during that game was not strictly greater than her best record at the time.

<b>Sample Input 1</b>

<code>     10
     3 4 21 36 10 28 35 5 24 42</code>
     
<b>Sample Output 1</b>

<code>     4 0</code>

<b>Explanation 1</b>

The diagram below depicts the number of times Maria broke her best and worst records throughout the season:



She broke her best record four times (after games $1$, $2$, $3$, and $9$) and her worst record zero times (no score during the season was lower than the one she earned during her first game), so we print 4 0 as our answer.

## Solution

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'breakingRecords' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts INTEGER_ARRAY scores as parameter.
#

def breakingRecords(scores):
    # Initialize count of Highs and Lows
    high_count, low_count = 0, 0
    
    # Initialize scores
    high, low = scores[0], scores[0]
    
    # Count new High or new Low in each iteration
    for item in scores:
        if item > high:
            high = item
            high_count += 1
        elif item < low:
            low = item
            low_count += 1
    
    print(high_count, low_count)

    
if __name__ == '__main__':
    
    # Input data
    n = int(input().strip())
    scores = list(map(int, input().rstrip().split()))
    
    # Call function
    breakingRecords(scores)
```
